```python
from sklearn.preprocessing import StandardScaler
col_names = ['Dinero', 'Consumo']
features = df[col_names]
scaler = StandardScaler().fit(features.values)
features = scaler.transform(features.values)
scaled_features = pd.DataFrame(features, columns = col_names)
scaled_features.head()
```

```python
gender = df['Genero']
newdf = scaled_features.join(gender)

newdf = pd.get_dummies(newdf, prefix=None, prefix_sep='_', dummy_na=False, columns=None, sparse=False, drop_first=False, dtype=None)

#newdf = newdf.drop(['Gender_Male'],axis=1)

newdf.head()
```

```python
Dispositivo = df['dispositivo']
newdf_2 = scaled_features.join(Dispositivo)

newdf_2 = pd.get_dummies(newdf_2, prefix=None, prefix_sep='_', dummy_na=False, columns=None, sparse=False, drop_first=False, dtype=None)

#newdf = newdf.drop(['Gender_Male'],axis=1)

newdf_2.head()
```

```python
from sklearn.cluster import KMeans

sse = {}
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(scaled_features)
    sse[k] = kmeans.inertia_ # SSE to closest cluster centroid

plt.title('The Elbow Method')
plt.xlabel('k')
plt.ylabel('SSE')
sns.pointplot(x=list(sse.keys()), y=list(sse.values()))
plt.show()
```

```python
model = KMeans(n_clusters=3, random_state=42)
model.fit(scaled_features)
model.labels_.shape
```

```python
df["Cluster"] = model.labels_
df.head()
```
